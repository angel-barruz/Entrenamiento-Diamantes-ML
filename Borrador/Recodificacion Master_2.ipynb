```python
# Librerías

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import datasets
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
import pickle
import joblib

from sklearn import linear_model
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import Ridge
from sklearn.svm import SVR
from sklearn.linear_model import SGDRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from sklearn.preprocessing import StandardScaler

from sklearn import preprocessing
```

```python
# CSV EDA
df=pd.read_csv('competi_kaggle.csv')
df
 
```

```python
df.info()
```

```python
X = df[['depth', 'table', 'x', 'y', 'z', 'color', 'clarity', 'city', 'cut', 'carat']]
X
```

```python
y = df[['price']]
y
```

```python
# One hot encoding
cat_cols = ['color', 'clarity', 'city', 'cut']
df_one_hot = pd.get_dummies(X, columns=cat_cols, drop_first=True)
df_one_hot
```

```python
# Scaling columnas numéricas
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_one_hot[['depth', 'table', 'x', 'y', 'z','carat']])
scaled_data
```

```python
df_X=pd.DataFrame(scaled_data)
df_X
```

```python
%%time

# Model definition

#model = LinearRegression()
model = RandomForestRegressor()

hyperparameters = model.get_params()



print(type(model), '\n')
print('Model hyperparameters:', hyperparameters, '\n')
```

```python
%%time

# Model training

model.fit(df_X, y)
#y_pred = model.predict(y)

print('Model:', model, '\n')
print('Model hyperparameters:', hyperparameters, '\n')
#print('Model coefficients:', model.coef_, '\n')
#print('Ground truth target:', y_test, '\n')
#print('Ground truth target:', y_pred, '\n')
```
